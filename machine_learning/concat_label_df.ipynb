```python
import os, re
import pandas as pd
data_path = './data_parquets/'

files = [file for file in os.listdir(data_path) if 'labeled_chunks' in file]
time_strs = [re.search(r'\d{4}-\d{2}-\d{2}_\d{6}', file).group() for file in files]
dfs = [pd.read_parquet(data_path+file) for file in files]
```

```python
new_dfs = []
for i, df in enumerate(dfs):
    length = df.shape[0]
    df = dfs[i].copy()
    df['begin_time'] = time_strs[i]
    df['sample_id'] = df.index
    new_dfs.append(df)

all_samples = pd.concat(new_dfs, ignore_index=True)
all_samples
```

```python
good = all_samples[all_samples['label'] == 'good']
bad = all_samples[all_samples['label'] == 'bad']
bad2good = all_samples[all_samples['label'] == 'bad2good']
print(good.shape[0], bad.shape[0], bad2good.shape[0])
```
