## 数据读取

```python
import numpy as np 
import pandas as pd

df = pd.read_csv('ecg_data-1110.csv')

def fake_data_stream():
    for idx in df.index:
        yield df.loc[idx, 'ecg']

#next(fake_data_stream())

def get_fake_idxs(fs=50, t=5):
    fake_idx_begin = np.random.randint(0, df.shape[0])
    fake_idxs = np.arange(fake_idx_begin, fake_idx_begin+t*fs)
    return fake_idxs
```

```python
import matplotlib.pyplot as plt
fake_idxs = get_fake_idxs(t=5)

plt.plot(fake_idxs, df.loc[fake_idxs, 'ecg'])
```
